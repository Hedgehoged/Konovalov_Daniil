```python
import math
import numpy as np
from plotly import express as px
from math import exp
```

```python
def f(x):
    return 0


def g(x):
    return 0
```

```python
N = 100    # Число точек решения
a, b = [0, 1]   
h = (b - a) / N    

xi = np.linspace(a - h, b + h, N + 2)    
ui = [1 for _ in xi]   
vi = [1 for _ in xi]   
fi = [f(e) for e in xi]
gi = [g(e) for e in xi]

K = 1000    # Максимальное число итераций
eps = 0.0001    # Точность решения
```

```python
def constructA():
    
    A = np.zeros((N + 2, N + 2))
    
    A[0, 1] = 1
    for i in range(1, N + 1):
        A[i, i - 1] = 1
        A[i, i] = -2
        A[i, i + 1] = 1
    A[N + 1, N - 1] = -1
    A[N + 1, N + 1] = 1

    return A


def constructB():
    
    B = np.zeros((N + 2, N + 2))
    
    B[0, 0] = 1
    B[0, 1] = -2
    B[0, 2] = 1

    for i in range(1, N + 1):
        B[i, i - 1] = -1
        B[i, i + 1] = 1
    B[N + 1, N] = 1

    return B
```

```python
def constructF(u, v):
 
    F = np.zeros(N + 2)
 
    F[0] = 0
    for i in range(1, N + 1):
        F[i] = u[i] * v[i] / (u[i]**2 + v[i]**2) - fi[i]
    F[N + 1] = 0

    return h * h * F


def constructG(u, v):
 
    G = np.zeros(N + 2)
 
    G[0] = 0
    for i in range(1, N + 1):
        G[i] = u[i] * v[i] / (u[i]**2 + v[i]**2) + gi[i]
    G[N + 1] = 0

    return 2 * h * G
```

```python
A = constructA()
B = constructB()

for i in range(K):    
    F = constructF(ui, vi)
    ui = np.linalg.solve(A, F)

    G = constructG(ui, vi)
    vi = np.linalg.solve(B, G)

    err = np.linalg.norm(np.concatenate((A.dot(ui) - constructF(ui, vi), B.dot(vi) - constructG(ui, vi))))
    if err < eps:
        print(f'Converged on {i + 1} iteration with tolerace {eps}. Error: {err:.5f}')
        break
else:
    print(f'Not converged after {i + 1} iteration. Error {err:.5f} exceeds tolerance {eps}')
```
